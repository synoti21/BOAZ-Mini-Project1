```python
import re
from tqdm import tqdm
from konlpy.tag import Okt
from collections import Counter
```

#### 정규표현식

```python
# 정규표현식을 사용하여 한글 표현만 남긴다
def korean(text):
    ko = re.compile('[^가-힣]')
    result = ko.sub(' ', str(text))
    return result
```

```python
# Counter를 사용하여 빈도수 확인
Counter(remove_stopwords).most_common()
# 전보다 훨씬 의미있는 단어들만 남았다
```

```python
# 너무 적게 나오는 단어들 제거
minimum_count = 3 # 최소 빈도수 = 3으로 지정
min_word=[]
for i in tqdm(range(len(remove_stopwords))):
    tmp = remove_stopwords[i]
    if remove_stopwords.count(tmp) >= minimum_count:
        min_word.append(tmp)
```

#### 전처리 함수

```python
# 전체 함수 생성
def make_wordlist(reviews,stopwords):
    words = korean(reviews)
    words = tokenizer.morphs(words,stem=True)
    words = [x for x in words if len(x)>1 or x==['양','맛','짱','잘']]
    words = [x for x in words if x not in stopwords]
    minimum_count = 3 # 최소 빈도수 = 3으로 지정
    final=[]
    for i in tqdm(range(len(words))):
        tmp = words[i]
        if words.count(tmp) >= minimum_count:
            final.append(tmp)
    return set(final)
```

```python
def preprocess(text, word_list):
    text = korean(text)
    tokenizer = Okt()
    text = tokenizer.morphs(text, stem=True)
    text = [x for x in text if x in word_list]
    return text
```

```python
word_list = make_wordlist(" ".join(map(str,df['리뷰'])),stopwords)
```

```python
# 전처리는 됐는데 ...
preprocess(df['리뷰'][43],word_list)
```
