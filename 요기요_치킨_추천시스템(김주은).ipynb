**자연어 처리**

```python
import re
from konlpy.tag import Okt
import konlpy
from konlpy.tag import Okt
import matplotlib.pyplot as plt
```

```python
okt=Okt()
```

```python
df['리뷰1'] = df['리뷰'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]","")
```

```python
df
```

```python
word_noun_list=[]

for word in df['리뷰1']:
  word_noun_list.extend(okt.morphs(word))
```

```python
word_noun_list_more=[]
multi_word = [w for w in word_noun_list if len(w) > 1]  
word_noun_list_more.extend(multi_word)
```

```python
from collections import Counter
word_counts = Counter(word_noun_list_more)
```

```python
word_max_num={}
for multi_word, counts in word_counts.most_common(30):
    word_max_num[multi_word] = counts
    print(f'{multi_word} : {counts}')
```

```python
frequent = Counter(word_noun_list_more).most_common()
print(frequent)
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(font_path='NanumGothic',background_color='ivory', width=800, height=600)
cloud = wordcloud.generate_from_frequencies(word_counts)
plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
